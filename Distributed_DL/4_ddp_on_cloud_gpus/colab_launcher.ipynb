# Distributed DL Unit 4 - Colab Launcher: `profile_ddp_gpu.py`

This notebook downloads the one file needed for the Unit 4 profiler path and launches it with `torchrun`.

Run it in a CUDA GPU runtime.

```python
import shlex
import subprocess
from pathlib import Path
import torch
from urllib.request import urlretrieve
def run(cmd: str) -> None:
    print(f"+ {cmd}", flush=True)
    subprocess.run(shlex.split(cmd), check=True)
```

```python
#Download the script:
urlretrieve("https://raw.githubusercontent.com/Idan-Alter/OU-22971-Toolbox/main/Distributed_DL/4_ddp_on_cloud_gpus/profile_ddp_gpu.py", "profile_ddp_gpu.py")
```

```python
has_gpu = torch.cuda.is_available() and torch.cuda.device_count() >= 1
print(f"has_gpu={has_gpu}")
if has_gpu:
    
    trace_name1 = "colab_gpu_initial"
    trace_dir   = "."
    launch_cmd1 = (
        "torchrun --standalone --nproc_per_node=1 "
        "profile_ddp_gpu.py "
        f"--trace-dir {trace_dir} "
        f"--trace-name {trace_name1} "
        "--num-workers 0 "
    )
    
    trace_name2 = "colab_gpu_baseline"
    launch_cmd2 = (
        "torchrun --standalone --nproc_per_node=1 "
        "profile_ddp_gpu.py "
        f"--trace-dir {trace_dir} "
        f"--trace-name {trace_name2} "
        "--num-workers 1 "
    )
    run(launch_cmd1)
    run(launch_cmd2)
    print("Trace and log files are in the cwd")
```
